### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda.columns
```

#1. Análisis de facturación



```python
lista_tienda=[]
lista_tienda2=[]
lista_tienda3=[]
lista_tienda4=[]

for x in tienda.Precio:
  lista_tienda.append(x)

for x in tienda2.Precio:
  lista_tienda2.append(x)

for x in tienda3.Precio:
  lista_tienda3.append(x)

for x in tienda4.Precio:
  lista_tienda4.append(x)

tienda_ingreso=sum(lista_tienda)
print("El ingreso de la tienda 1 es:",tienda_ingreso)

tienda2_ingreso=sum(lista_tienda2)
print("El ingreso de la tienda 2 es:",tienda2_ingreso)

tienda3_ingreso=sum(lista_tienda3)
print("El ingreso de la tienda 3 es:",tienda3_ingreso)

tienda4_ingreso=sum(lista_tienda4)
print("El ingreso de la tienda 4 es:",tienda4_ingreso)
```

# 2. Ventas por categoría

```python
categoria_tienda=[]
categoria_tienda2=[]
categoria_tienda3=[]
categoria_tienda4=[]

for x in tienda ["Categoría del Producto"]:
  categoria_tienda.append(x)

for x in tienda2 ["Categoría del Producto"]:
  categoria_tienda2.append(x)

for x in tienda3 ["Categoría del Producto"]:
  categoria_tienda3.append(x)

for x in tienda4 ["Categoría del Producto"]:
  categoria_tienda4.append(x)

ventas_categoria_tienda=dict(zip(categoria_tienda,map(lambda x: categoria_tienda.count(x),categoria_tienda)))
ventas_categoria_tienda2=dict(zip(categoria_tienda2,map(lambda x: categoria_tienda2.count(x),categoria_tienda2)))
ventas_categoria_tienda3=dict(zip(categoria_tienda3,map(lambda x: categoria_tienda3.count(x),categoria_tienda3)))
ventas_categoria_tienda4=dict(zip(categoria_tienda4,map(lambda x: categoria_tienda4.count(x),categoria_tienda4)))

print("Ventas por categoría de la tienda 1:",ventas_categoria_tienda)
print("Ventas por categoría de la tienda 2:",ventas_categoria_tienda2)
print("Ventas por categoría de la tienda 3:",ventas_categoria_tienda3)
print("Ventas por categoría de la tienda 4:",ventas_categoria_tienda4)

```

# 3. Calificación promedio de la tienda


```python
promedio_tienda=[]
promedio_tienda2=[]
promedio_tienda3=[]
promedio_tienda4=[]

for x in  tienda["Calificación"]:
  promedio_tienda.append(x)

for x in  tienda2["Calificación"]:
  promedio_tienda2.append(x)

for x in  tienda3["Calificación"]:
  promedio_tienda3.append(x)

for x in  tienda4["Calificación"]:
  promedio_tienda4.append(x)

calificacion_tienda=sum(promedio_tienda)/len(promedio_tienda)
calificacion_tienda2=sum(promedio_tienda2)/len(promedio_tienda2)
calificacion_tienda3=sum(promedio_tienda3)/len(promedio_tienda3)
calificacion_tienda4=sum(promedio_tienda4)/len(promedio_tienda4)

print("La calificación promedio de la tienda 1:",calificacion_tienda)
print("La calificación promedio de la tienda 2:",calificacion_tienda2)
print("La calificación promedio de la tienda 3:",calificacion_tienda3)
print("La calificación promedio de la tienda 4:",calificacion_tienda4)

```

# 4. Productos más y menos vendidos

```python
lista_producto_tienda=[]
lista_producto_tienda2=[]
lista_producto_tienda3=[]
lista_producto_tienda4=[]

for x in tienda["Producto"]:
  lista_producto_tienda.append(x)

for x in tienda2["Producto"]:
  lista_producto_tienda2.append(x)

for x in tienda3["Producto"]:
  lista_producto_tienda3.append(x)

for x in tienda4["Producto"]:
   lista_producto_tienda4.append(x)

ventas_producto_tienda=dict(zip(lista_producto_tienda,map(lambda x: lista_producto_tienda.count(x),lista_producto_tienda)))
ventas_producto_tienda2=dict(zip(lista_producto_tienda2,map(lambda x: lista_producto_tienda2.count(x),lista_producto_tienda2)))
ventas_producto_tienda3=dict(zip(lista_producto_tienda3,map(lambda x: lista_producto_tienda3.count(x),lista_producto_tienda3)))
ventas_producto_tienda4=dict(zip(lista_producto_tienda4,map(lambda x: lista_producto_tienda4.count(x),lista_producto_tienda4)))

ventas_productos_ordenados_tienda = dict(sorted(ventas_producto_tienda.items(), key=lambda item: item[1], reverse=True))
ventas_productos_ordenados_tienda2 = dict(sorted(ventas_producto_tienda2.items(), key=lambda item: item[1], reverse=True))
ventas_productos_ordenados_tienda3 = dict(sorted(ventas_producto_tienda3.items(), key=lambda item: item[1], reverse=True))
ventas_productos_ordenados_tienda4 = dict(sorted(ventas_producto_tienda4.items(), key=lambda item: item[1], reverse=True))
print("Productos ordenados por cantidad de ventas:",ventas_productos_ordenados_tienda)
print("Productos ordenados por cantidad de ventas:",ventas_productos_ordenados_tienda2)
print("Productos ordenados por cantidad de ventas:",ventas_productos_ordenados_tienda3)
print("Productos ordenados por cantidad de ventas:",ventas_productos_ordenados_tienda4)
```

# 5. Envío promedio por tienda

```python
lista_costo_tienda=[]
lista_costo_tienda2=[]
lista_costo_tienda3=[]
lista_costo_tienda4=[]

for x in tienda["Costo de envío"]:
  lista_costo_tienda.append(x)

for x in tienda2["Costo de envío"]:
  lista_costo_tienda2.append(x)

for x in tienda3["Costo de envío"]:
  lista_costo_tienda3.append(x)

for x in tienda4["Costo de envío"]:
  lista_costo_tienda4.append(x)

promedio_costo_tienda=sum(lista_costo_tienda)/len(lista_costo_tienda)
promedio_costo_tienda2=sum(lista_costo_tienda2)/len(lista_costo_tienda2)
promedio_costo_tienda3=sum(lista_costo_tienda3)/len(lista_costo_tienda3)
promedio_costo_tienda4=sum(lista_costo_tienda4)/len(lista_costo_tienda4)

print("El costo envío promedio de la tienda 1 es:",promedio_costo_tienda)
print("El costo envío promedio de la tienda 2 es:",promedio_costo_tienda2)
print("El costo envío promedio de la tienda 3 es:",promedio_costo_tienda3)
print("El costo envío promedio de la tienda 4 es:",promedio_costo_tienda4)
```
